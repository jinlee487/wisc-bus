```python
#creating weekday_stop_time_fix
#open the stop_times table from the mmt_gtfs folder
stop_times =  pd.read_csv(r"C:\Users\이진우\cs_summer\Jin_project\mmt_gtfs\stop_times.txt", sep=",", engine='python')
stop_times.head()
```

```python
# route_short_name == the bus number
# the trip_id can be linked with the trip id in stop times
# trip id is unique to each row. 
trips =  pd.read_csv(r'C:\\Users\\이진우\\cs_summer\\Jin_project\\mmt_gtfs\\trips.txt', sep=",", engine='python')
trips.head()
```

```python
#trip_id is the id of each bus trip. throughout the day, multiple #1 bus makes a trip at different times. 
#route_id 
#take only the WKD for service id.
trips_wkd = []
for index, row in trips.iterrows():
    if 'WKD' in row['service_id']:
        trips_wkd.append(row) 
trips_wkd=pd.DataFrame(trips_wkd)
```

```python
#what is MWK???? will just ignore...
uniq=trips_wkd['trip_id'].unique()
only_uniq_trip_id=trips_wkd.groupby(by=['trip_id'], as_index=False).first()
```

```python
#got rid of the : in the times so i can turn them in to int
#only the weekday buses
stop_times_time_fix = copy.deepcopy(stop_times)
stop_times_time_fix['arrival_time_:'] = (stop_times_time_fix['arrival_time'].str.replace(':', '')).astype(int)
stop_times_time_fix['departure_time_:'] = (stop_times_time_fix['departure_time'].str.replace(':', '')).astype(int)
list_of_trip_id = []

for u in uniq:
    list_of_trip_id.append(u)

weekday_stop_time_fix = stop_times_time_fix[stop_times_time_fix['trip_id'].isin(list_of_trip_id)]
```
