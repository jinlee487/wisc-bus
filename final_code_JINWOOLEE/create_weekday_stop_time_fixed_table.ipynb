```python
#stop_times table only weekday schedules
weekday_stop_time_fix =  pd.read_csv(r'weekday_stop_time_fix.csv', sep=",", engine='python')
weekday_stop_time_fix.head()
```

```python
weekday_stop_time_fix['time_diff']=''
weekday_stop_time_fix['seq'] = ''
weekday_stop_time_fix['total_time'] = 0
list_of_stops = weekday_stop_time_fix['stop_id'].unique()
```

```python
weekday_stop_time_fix['arrival_time_fix'] = (weekday_stop_time_fix['arrival_time'].str.replace(':', '')).astype(int)
weekday_stop_time_fix['departure_time_fix'] = (weekday_stop_time_fix['departure_time'].str.replace(':', '')).astype(int)
```

```python
#This secotion will take a long time(maybe an hour or longer). It iterates every single row 
list_of_time = [] 
for index, row in weekday_stop_time_fix.iterrows():
    fixed_time_a = row['arrival_time']
    fixed_time_a = fixed_time_a.split(':')
    fixed_time_a = int(fixed_time_a[0])*60*60+int(fixed_time_a[1])*60+int(fixed_time_a[2])
    weekday_stop_time_fix.loc[index,'arrival_time_fix']= fixed_time_a
    
    fixed_time_d = row['departure_time']
    fixed_time_d = fixed_time_d.split(':')
    fixed_time_d = int(fixed_time_d[0])*60*60+int(fixed_time_d[1])*60+int(fixed_time_d[2])
    weekday_stop_time_fix.loc[index,'departure_time_fix']= fixed_time_d
    
#     list_of_time.append(fixed_time)
# pd.to_datetime(list_of_time).time
```
